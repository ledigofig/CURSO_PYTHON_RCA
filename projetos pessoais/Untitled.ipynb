```python
#carregando latitude e ligitude dos municipios de MT
url_cordendadas = 'https://raw.githubusercontent.com/alanwillms/geoinfo/master/latitude-longitude-cidades.csv'
df_cordenadas = pd.read_csv(url_cordendadas, sep=';', skiprows=0)
df_cordenadas
```

```python
df_cordenadas = df_cordenadas[df_cordenadas['uf'] == 'MT'] #separando dados dos municipios de MT
df_cordenadas = df_cordenadas[['municipio','longitude','latitude']] #pegando somente os dados necessarios para plotar o mapa
df_cordenadas.head()#mostrando dados
```

```python
df = pd.read_csv("covid-26-02-13-03 - Sheet1 v3.csv")
df
```

```python
#inserindo indice como municipio
df_cidade_corona = df.copy()#fazendo copia do dataframe
df_cidade_corona.set_index('Municipio',inplace=True,drop=False)
df_cidade_corona
```

```python
#Definindo indice pela cidade 
df_cordenadas.set_index('municipio',inplace=True)
df_cordenadas.head()
```

```python
#pegando nome de cidades e unindo com codigo base no nome
df_cidade_corona_cordenada = df_cidade_corona.join(df_cordenadas)

df_cidade_corona_cordenada
```

```python
#Definindo o numero como index
df_cidade_corona_cordenada.set_index('numero',inplace=True,)
df_cidade_corona_cordenada
```

```python
for (i,linha) in df_cidade_corona_cordenada.iterrows():
    #cod = row['Municipio'+'latitude'='longitude']
    folium.Marker(
        location=[linha['latitude'],linha['longitude']],
        popup='teste',
        #icon=folium.Icon(color='green')
        tooltip= ['Municipio']
    ).add_to(mapa_mt)
    #print(cod)
    #print(str(linha['Municipio'])
    
#fonte do for
#: https://medium.com/capivarapython/itera%C3%A7%C3%A3o-em-pandas-dataframes-e-desempenho-72d2d12522e1
```
