```python
import pandas as pd
```

```python
#carregando latitude e ligitude dos municipios de MT
url_cordendadas = 'https://raw.githubusercontent.com/alanwillms/geoinfo/master/latitude-longitude-cidades.csv'
df_cordenadas = pd.read_csv(url_cordendadas, sep=';', skiprows=0)
df_cordenadas
```

```python
df_cordenadas = df_cordenadas[df_cordenadas['uf'] == 'MT'] #separando dados dos municipios de MT
df_cordenadas = df_cordenadas[['municipio','longitude','latitude']] #pegando somente os dados necessarios para plotar o mapa
df_cordenadas.head()#mostrando dados
```

## Fazendo a união dos dados

```python
#inserindo indice como municipio
df_cidade_corona = df.copy()#fazendo copia do dataframe
df_cidade_corona.set_index('Município',inplace=True,drop=False)
df_cidade_corona
```

```python
#Definindo indice pela cidade 
df_cordenadas.set_index('municipio',inplace=True,drop=False)
df_cordenadas.head()
```

```python
#pegando nome de cidades e unindo com base no nome
df_cidade_corona_cordenada = df_cidade_corona.join(df_cordenadas)

df_cidade_corona_cordenada
```

```python
df_cidade_corona_cordenada.head(60)
```

```python
df_cidade_corona_cordenada.set_index('id_municipio',inplace=True,drop=False)
df_cidade_corona_cordenada
```

```python
df_cidade_corona_cordenada.update(df_cidade_corona_cordenada['longitude'].fillna(str(0)))
df_cidade_corona_cordenada.update(df_cidade_corona_cordenada['latitude'].fillna(str(0)))
```

```python
#TRATANDO DADOS NaN

#para alteração
df_cidade_corona_cordenada.update(df_cidade_corona_cordenada['CONFIRMADO'].fillna(str(0)))
df_cidade_corona_cordenada.update(df_cidade_corona_cordenada['SUSPEITO'].fillna(str(0)))
df_cidade_corona_cordenada.update(df_cidade_corona_cordenada['DESCARTADO'].fillna(str(0)))
#df_cidade_corona_cordenada.update(df_cidade_corona_cordenada['EXCLUÍDOS'].fillna(str(0)))

df_cidade_corona_cordenada.to_html('arquivo.html')
```

```python
from folium.plugins import heat_map
from folium import plugins
import os


mcg = folium.plugins.MarkerCluster(control=False)
mapa_mt_calor.add_child(mcg)


lat = df_cidade_corona_cordenada.latitude.to_list()
lng = df_cidade_corona_cordenada.longitude.to_list()

mapa_mt_calor = folium.Map(
    location=[-12.6470066,-58.4244261],
    name="Mapa de Calor",
    tiles='Stamen Toner',
    zoom_start=6,
    control_scale=True,
    prefer_canvas=True)

HeatMap(list(zip(lat, lng))).add_to(mapa_mt_calor)



folium.LayerControl(collapsed=False).add_to(mapa_mt_calor)
mapa_mt_calor.save('calor.html')
mapa_mt_calor

g2 = folium.plugins.FeatureGroupSubGroup(mcg, 'group2')
mapa_mt_calor.add_child(g2)

# -*- coding: utf-8-*-
for (i,linha) in df_cidade_corona_cordenada.iterrows():
    #cod = row['Municipio'+'latitude'='longitude']
    folium.map.Marker(
      [linha['latitude'], linha['longitude']],
        #tooltip = [i],#tinha que ter colocado o indice
        popup=('<b bgcolor="#FF0000">Confirmado </b>' +str(linha['CONFIRMADO'])+'<br>'+
               'Suspeito ' +str(linha['SUSPEITO'])+'<br>'+
               'Descartado ' +str(linha['DESCARTADO'])+'<br>'
              ),
        icon=folium.Icon(color='orange',icon='ambulance',prefix='fa')
        #tooltip="suspeito"
    ).add_to(mapa_mt_calor)
    #print(cod)
    #print(linha['Suspeito'])
    #print(linha['longitude'])
    
folium.LayerControl(collapsed=False).add_to(mapa_mt_calor)
mapa_mt_calor.save('mapa.html')

plugins.Fullscreen(
    position='topright',
    title='Expand me',
    title_cancel='Exit me',
    force_separate_button=True
).add_to(mapa_mt_calor)

folium.LayerControl(collapsed=False).add_to(mapa_mt_calor)
mapa_mt_calor  
# fonte do for
#: https://medium.com/capivarapython/itera%C3%A7%C3%A3o-em-pandas-dataframes-e-desempenho-72d2d12522e1
```

# Plotando o mapa com if e cor vermelha nos municipios confirmados

```python
import branca
mapa_cor_mt = folium.Map(
    location=[-12.6470066,-58.4244261]
    ,zoom_start=6)
for (i, linha) in df_cidade_corona_cordenada.iterrows():
    #t=str(linha["Municipio"])
    #cod = row['Municipio'+'latitude'='longitude']
    if int(linha['CONFIRMADO']) > 0:
         folium.Marker(
          [linha['latitude'], linha['longitude']],
            tooltip = 'Veja mais informaçoes',#tinha que ter colocado o indice
            popup=('<b bgcolor="#FF0000">Confirmado </b>' +str(linha['CONFIRMADO'])+'<br>'+
                   'Suspeito ' +str(linha['SUSPEITO'])+'<br>'+
                   'Descartado ' +str(linha['DESCARTADO'])+'<br>'
                    ),
            icon=folium.Icon(color='red',icon='ambulance',prefix='fa')
        #tooltip="suspeito"
    ).add_to(mapa_cor_mt)
    elif int(linha['SUSPEITO']) > 0:
         folium.Marker(
          [linha['latitude'], linha['longitude']],
            tooltip = 'Veja mais informaçoes',#tinha que ter colocado o indice
            popup=('<b bgcolor="#FF0000">Confirmado </b>' +str(linha['CONFIRMADO'])+'<br>'+
                   'Suspeito ' +str(linha['SUSPEITO'])+'<br>'+
                   'Descartado ' +str(linha['DESCARTADO'])+'<br>'
                    ),
            icon=folium.Icon(color='orange',icon='ambulance',prefix='fa')
        #tooltip="suspeito"
    ).add_to(mapa_cor_mt)
    else:
                folium.Marker(
                              [linha['latitude'], linha['longitude']],
                                tooltip = 'Caso excluidos de corona virus',#tinha que ter colocado o indice
                                popup=('<b bgcolor="#FF0000">Confirmado </b>' +str(linha['CONFIRMADO'])+'<br>'+
                                'Suspeito ' +str(linha['SUSPEITO'])+'<br>'+
                                'Descartado ' +str(linha['DESCARTADO'])+'<br>'
                            ),
                    
                icon=folium.Icon(color='green',icon='thumbs-o-up',prefix='fa')
            #tooltip="suspeito"
        ).add_to(mapa_cor_mt)
            
   
    #print(cod)
    #print(linha['Suspeito'])
    #print(linha['longitude'])

    
mapa_cor_mt
```
